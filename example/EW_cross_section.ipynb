```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
plt.rcParams['font.size'] = 14
mmin = 0
mmax = 2.5
```

```python
df0 = pd.read_csv('input_mag.in', delimiter='\t', names=('y', 'x', 'z', 'f'))

x0 = df0.x.values
y0 = df0.y.values
z0 = df0.z.values
f0 = df0.f.values

x00 = np.unique(x0)
y00 = np.unique(y0)

nx0 = len(x00)
ny0 = len(y00)

xx0 = x0.reshape(ny0, nx0)
yy0 = y0.reshape(ny0, nx0)
ff0 = f0.reshape(ny0, nx0)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
pc = ax.pcolor(yy0, xx0, ff0, cmap='jet')
ax.contour(yy0, xx0, ff0, levels=20)
ax.set_aspect('equal')
ax.set_xlabel('y(km)')
ax.set_ylabel('x(km)')
ax.set_title('Input Anomaly')
plt.colorbar(pc, orientation='horizontal', shrink=0.5, label='(nT)', pad=0.125)
```

```python
def cross_section(fn, xc):
    df = pd.read_csv(fn, delimiter='\t', names=('y', 'x', 'z', 'b'))
    x = df.x.values
    y = df.y.values
    z = df.z.values
    b = df.b.values

    x1 = np.unique(x)
    y1 = np.unique(y)
    z1 = np.unique(z)

    nx = len(x1)
    ny = len(y1)
    nz = len(z1)

    # x(EW)-z cross section
    dx = x1[2] - x1[1]
    dy = y1[2] - y1[1]
    xc0 = 0.0 + dx / 2.
    idx = np.where(np.abs(x - xc0) < dx / 2.)[0]

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    pc = ax.pcolor(y[idx].reshape(nz, nx), -z[idx].reshape(nz, nx), b[idx].reshape(nz, nx), cmap='jet')
    pc.set_clim(mmin, mmax)
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(2.5, 0)
    ax.set_xlabel('y(km)')
    ax.set_ylabel('z(km)')
    ax.set_aspect('equal')
    ax.set_title('True model: East-West Cross-Section: x = {:.2f} km'.format(xc0))
    plt.colorbar(pc, orientation='horizontal', shrink=0.8, label='(A/m)', pad=0.125)
```

```python
cross_section('true.model', 0.05)
cross_section('model.data', 0.05)
```
